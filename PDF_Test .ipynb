Below are 2 normally distributed time series:

```python
data1, data2 = np.random.randn(365, 1)+2, np.random.randn(365, 1)-2
data = np.concatenate((data1,data2), axis=1)
start, end = datetime.datetime(2016, 1, 1), datetime.datetime(2016, 12, 30)
test = pd.DataFrame(data=data,
                    index=pd.DatetimeIndex(start=start, end=end, freq="D"),
                    columns=["1", "2"],
                    dtype=float)
test.plot()
plt.show()
```
